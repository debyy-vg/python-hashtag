```python
import pandas as pd
    
nubank_df = pd.read_csv(r'./nubank-2021-05.csv')    
display(nubank_df)    
    
```

```python
# tqdm #Saber quanto tempo esta demorando seu codigo
import time
from tqdm import tqdm

#position e leave coloca tudo em uma brra só 
pbar = tqdm(total=len(nubank_df['amount']), position = 1, leave= True)
            

for i, categoria in enumerate(nubank_df['amount']):
    pbar.update()
    if categoria == 110.00:
        nubank_df.loc[i, 'category',] = 'ANULAR'
        time.sleep(0.4)
display(nubank_df)        

        
        
        
 

```
