#### Importanto arquivo e renomeando coluna:

```python
import pandas as pd

nubank_df = pd.read_csv(r'./nubank-2021-05.csv')
#Renomear coluna
nubank_df = nubank_df.rename(columns={'amount': 'Valores'})

nubank_df = nubank_df.rename(columns={'category': 'Categoria'})
nubank_df = nubank_df.rename(columns={'date': 'Data da Compra'})
nubank_df


```

#### Agrupando valores e plotando gráfico:

```python
categorias_df = nubank_df.groupby('Categoria').sum()
categorias_df.plot(figsize = (15,5),  kind='barh', color = 'pink')
display(categorias_df)
```

##### Somando Valores:

```python
qtdd_total = nubank_df['Valores'].sum()
display(qtdd_total)
```

#### Filtrando valores:

```python
gasto_mercado = nubank_df[nubank_df['Categoria'] == 'supermercado']
gasto_mercado
```

#### Filtrar valores com mais parametros :

```python
compras_enxuto = nubank_df[(nubank_df['Categoria'] == 'supermercado') & (nubank_df['title'] == 'Enxuto Lj 08 Campinas')]
display(compras_enxuto)
```

#### Alterando data:


```python
#alterar formata data
nubank_df['Data da Compra'] = pd.to_datetime(nubank_df['Data da Compra'], format='%Y-%m-%d')
nubank_df['Ano da Venda'] = nubank_df['Data da Compra'].dt.year
nubank_df['Mês da Venda'] = nubank_df['Data da Compra'].dt.month
nubank_df['Dia da Venda'] = nubank_df['Data da Compra'].dt.day
nubank_df
```

### Alterando valor na dataframe utilizando loc e iloc

```python
#head demostra apenas os 5 primeiros numeros 
# set.index = altera o o nome para on indice do produto

nubank_df = nubank_df.set_index('Categoria')
display(nubank_df.head())


###LOC

nubank_df.loc['supermercado', 'title']= 'Gastos com mercado'
display(nubank_df)




```
