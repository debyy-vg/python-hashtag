### Importando tabelas:

```python
import pandas as pd

funcionario_df = pd.read_csv('./CadastroFuncionarios.csv', sep = ';', decimal = ',')
faturamento_df = pd.read_excel('./BaseServiçosPrestados.xlsx')
meses_df = pd.read_csv('./CadastroClientes.csv', sep = ';')
display(faturamento_df)
display(meses_df)
display(funcionario_df)



```

### 1- Valor Total da Folha Salarial:

```python
funcionario_df = funcionario_df.drop(columns=['Estado Civil','Nome Completo','Cargo'])
funcionario_df['Salario total'] = funcionario_df['Salario Base']+ funcionario_df['Impostos']+ funcionario_df['Beneficios'] + funcionario_df['VT'] + funcionario_df['VR']
display(funcionario_df)


print('A Folha Salarial da empresa é R$ {:,}'.format(funcionario_df['Salario total'].sum()))
```

### 2 - Qual foi o faturamento da empresa?
Sugestão: calcule o faturamento total de cada serviço e depois some o faturamento de todos







```python
#importa planilha


#juntas duas colunas das planilhas com merge:
total_df = faturamento_df[['ID Cliente','Tempo Total de Contrato (Meses)']].merge(meses_df[['ID Cliente', 'Valor Contrato Mensal']], on ='ID Cliente')
total_df['Faturamento Total'] = total_df['Tempo Total de Contrato (Meses)']* total_df['Valor Contrato Mensal'] 

#soma faturamento:   
display(total_df['Faturamento Total'].sum())
print('Faturamento total é de R${:,}'.format(sum(total_df['Faturamento Total'])))
```

### 3 - Qual o % de funcionários que já fechou algum contrato?

```python
# .unique remove valores duplicados 
qtde_funcionario_fecharamcontrato = len(faturamento_df['ID Funcionário'].unique())
qtde_funcionario_total = len(funcionario_df['ID Funcionário'])
print('Percentual Funcionários Fecharam contratos:{:.2%}'.format(qtde_funcionario_fecharamcontrato/qtde_funcionario_total))
```

### 4 - Calcule o total de contratos que cada área da empresa já fechou:

```python
contratos_area_df = faturamento_df[['ID Funcionário']].merge(funcionario_df[['ID Funcionário', 'Area']] , on = 'ID Funcionário')
contratos_area_qtde = contratos_area_df['Area'].value_counts()
contratos_area_qtde.plot(kind = 'bar', color= 'red')
```

### 5 - Calcule o total de funcionários por área:

```python
funcionarios_areas = funcionario_df['Area'].value_counts()
funcionarios_areas
funcionarios_areas.plot(kind = 'pie')
```

### 6 -Qual o ticket médio mensal (faturamento médio mensal) dos contratos?


```python
ticket_medio = meses_df['Valor Contrato Mensal'].mean()
print('Ticket Médio Mensal: R${:.2f}'.format(ticket_medio))
```
